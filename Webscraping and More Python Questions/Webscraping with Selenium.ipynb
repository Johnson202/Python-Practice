```python
# documentation: https://www.selenium.dev/selenium/docs/api/py/webdriver/selenium.webdriver.common.by.html
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
#from selenium.webdriver.common.by import By
import pandas as pd
```

```python
#!pip install selenium
```

```python
website = 'https://www.usinflationcalculator.com/inflation/historical-inflation-rates/'
```

```python
service = Service()
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=service, options=options)
```

```python
driver.get(website)
```

```python
# XPath for buttons, as applicable //tagName[@AttributeName='Value']
#xpath = ...
#button = driver.find_element_by_xpath(xpath)
#button.click()
```

```python
data = driver.find_elements('tag name', 'tr') #can also use 'tag name'
```

```python
for datum in data:
    print(datum.text)
    print(datum.text.split(" "))
```

```python
year = []
january = []
february = []
march = []
april = []
may = []
june = []
july = []
august = []
september = []
october = []
november = []
december = []
average = []
```

```python
for datum in data:
    lst_datum = datum.text.split(" ")
    year.append(lst_datum[0])
    january.append(lst_datum[1])
    february.append(lst_datum[2])
    march.append(lst_datum[3])
    april.append(lst_datum[4])
    may.append(lst_datum[5])
    june.append(lst_datum[6])
    july.append(lst_datum[7])
    august.append(lst_datum[8])
    september.append(lst_datum[9])
    october.append(lst_datum[10])
    november.append(lst_datum[11])
    december.append(lst_datum[12])
    average.append(lst_datum[13])
```

```python
dict_data = {year[0]: year[1:], january[0]: january[1:], february[0]: february[1:], march[0]: march[1:], 
             april[0]: april[1:], may[0]: may[1:], june[0]: june[1:], july[0]: july[1:], august[0]: august[1:],
             september[0]: september[1:], october[0]: october[1:], november[0]: november[1:], december[0]: december[1:],
             average[0]: average[1:]}
```

```python
dict_data
```

```python
df = pd.DataFrame(dict_data)
```

```python
df
```

```python
df['Dec'][109] = float('NaN')
df['Ave'][109] = float('NaN')
```
