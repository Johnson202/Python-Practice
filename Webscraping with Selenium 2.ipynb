```python
# documentation: https://www.selenium.dev/selenium/docs/api/py/webdriver/selenium.webdriver.common.by.html
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
#from selenium.webdriver.common.by import By
import pandas as pd
```

```python
#!pip install selenium
```

```python
website = 'https://www.thebalancemoney.com/unemployment-rate-by-year-3305506'
```

```python
service = Service()
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=service, options=options)
```

```python
driver.get(website)
```

```python
# XPath for buttons, as applicable //tagName[@AttributeName='Value']
#xpath = ...
#button = driver.find_element_by_xpath(xpath)
#button.click()
```

```python
data = driver.find_elements('tag name', 'tr') #can also use 'tag name'
```

```python
for datum in data:
    print(datum.text)
    print(datum.text.split(" "))
```

```python
year = []
unemployment_rate = []
annual_gdp_growth = []
inflation_rate = []
notable_events = []
```

```python
for i in range(1, len(data)):
    datum = data[i]
    lst_datum = datum.text.split(" ")
    year.append(lst_datum[0])
    unemployment_rate.append(lst_datum[1])
    annual_gdp_growth.append(lst_datum[2])
    inflation_rate.append(lst_datum[3])
    notable_events.append(' '.join(lst_datum[4:]))
```

```python
dict_data = {"year": year, "unemployment_rate": unemployment_rate, "annual_gdp_growth": annual_gdp_growth,
             "inflation_rate": inflation_rate, "notable_events": notable_events}
```

```python
dict_data
```

```python
df = pd.DataFrame(dict_data)
```

```python
df
```

```python
df['unemployment_rate'] = df.unemployment_rate.apply(lambda x: float(x[:-1])*.01)
```

```python
df['inflation_rate'] = df.inflation_rate.apply(lambda x: float(x[:-1])*.01)
```

```python
df[df['annual_gdp_growth'] == "NA"]
```

```python
df['annual_gdp_growth'] = df.annual_gdp_growth.apply(lambda x: float('NaN') if x == "NA" else float(x[:-1])*.01)
```

```python
df['year'] = df['year'].astype(int)
```

```python
df.to_csv("./economic_data.csv", index=False)
```
